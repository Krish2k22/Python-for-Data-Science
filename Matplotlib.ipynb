## Matplotlib

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

### Line Plot

```python
x = np.arange(1, 11)
x
```

```python
y = x * 2
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
plt.plot(x, y, color ="green", linestyle = ":", linewidth = 2)
plt.title("X vs Y")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.show()
```

```python
x = np.arange(1, 11)
y1 = x * 2
y2 = x * 3
```

```python
plt.plot(x, y1, color = "red", linestyle = ":", linewidth = 2)
plt.plot(x, y2, color = "blue", linewidth = 2)
plt.grid(True)
plt.title("Two lines in a plot")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.show()
```

```python
x = np.arange(1, 11)
y1 = x * 2
y2 = x * 3
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y1, color = "red", linestyle = ":", linewidth  =2)
plt.show()

plt.subplot(2, 1, 2)
plt.plot(x, y2, color = "blue", linewidth = 2 )
plt.title("Two lines in a plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

### Bar Plot

```python
student = {"Krish": 96, "Abir": 87, "Anirban": 78}
names = list(student.keys())
marks = list(student.values())
```

```python
plt.bar(names, marks)
plt.title("Distribution of Student Marks")
plt.xlabel("Name of Students")
plt.ylabel("Marks os Students")
plt.show()
```

```python
plt.barh(names, marks, color = "green") 
plt.title("Distribution of Student Marks")
plt.xlabel("Marks os Students")
plt.ylabel("Name of Students")
plt.show()
```

### Scatter Plot

```python
x = np.random.randint(0, 50, 10)
y = np.random.randint(51, 100, 10)
```

```python
print(x)
print(y)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y, marker = "*", color = "green", s = 100)
plt.show()
```

```python
x = np.array([10, 20, 53, 15, 19, 38, 45, 59])
y = np.array([22, 54, 19, 36, 47, 9, 55, 60])
z = np.array([12, 24, 36, 45, 7, 58, 64, 78])
```

```python
plt.scatter(x, y)
plt.scatter(x, z, marker = "*", color = "orange", s = 100)
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.scatter(x, y, color = "red")
plt.title("1st Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(1, 2, 2)
plt.scatter(x, z, color = "blue", marker = "*", s = 100)
plt.title("2nd Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

### Box-Plot

```python
one = np.random.randint(1, 11, 8)
two = np.random.randint(1, 11, 8)
three = np.random.randint(1, 11, 8)
one
```

```python
print(np.sort(one))
print(np.sort(two))
print(np.sort(three))
```

```python
data = list([one, two, three])
data
```

```python
plt.boxplot(data)
plt.show()
```

### Violin-Plot

```python
plt.violinplot(data, showmedians = True)
plt.show()
```

### Pie-Chart

```python
fruits = ["Apple", "Mango", "Orange", "Guava"]
quantity = [67, 34, 100, 29]
```

```python
plt.pie(quantity, labels = fruits)
plt.show()
```

```python
plt.pie(quantity, labels = fruits, autopct = "%0.1f%%", colors = ["red", "yellow", "orange", "green"])
plt.show()
```

### DoughNut-Chart

```python
fruits = ["Apple", "Mango", "Orange", "Guava"]
quantity = [67, 34, 100, 29]
```

```python
plt.pie(quantity, labels = fruits, colors = ["red", "yellow", "orange", "green"], radius = 1)
plt.pie([1], radius = 0.5, colors = "white")
plt.show()
```
